```python
df.Start_Time
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'], format='mixed', errors='coerce')
```

```python
sns.distplot(df.Start_Time.dt.hour,norm_hist = True,bins=24,kde=False)
```

```python
sns.distplot(df.Start_Time.dt.dayofweek,norm_hist = True,bins=7,kde=False)
```

```python
sunday_sart_time = df.Start_Time[df.Start_Time.dt.dayofweek==6]
sns.distplot(sunday_sart_time.dt.hour,norm_hist = True,bins=24,kde=False)
```

```python
monday_sart_time = df.Start_Time[df.Start_Time.dt.dayofweek==0]
sns.distplot(monday_sart_time.dt.hour,norm_hist = True,bins=24,kde=False)
```

```python
import folium
from folium.plugins import HeatMap
zip(list(df.Start_Lat), list(df.Start_Lng))
sample_df = df.sample(int(0.001 * len(df)))
lat_lon_pairs = list(zip(list(sample_df.Start_Lat), list(sample_df.Start_Lng)))
map = folium.Map()
HeatMap(lat_lon_pairs).add_to(map)
map
```
